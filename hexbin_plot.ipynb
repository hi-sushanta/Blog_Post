```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.random.normal(loc=5, scale=10, size=5000)
y = np.random.normal(loc=10, scale=15, size=5000)

x = np.concatenate((x, np.random.normal(loc=-5, scale=10, size=5000)))
y = np.concatenate((y, np.random.normal(loc=-10, scale=15, size=5000)))

# Create a hexbin plot
plt.figure(figsize=(10, 7))
plt.hexbin(x, y, gridsize=40, cmap='Greens')
plt.colorbar(label='Count in bin')
plt.title('Hexagonal Binning Plot')

# remove spine
plt.gca().spines[['top', 'right']].set_visible(False)
```
